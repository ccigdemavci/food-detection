```python
# 📦 1. Gerekli kütüphaneleri yükle
!pip install ultralytics --upgrade

# 🚀 2. YOLOv8'i içe aktar
from ultralytics import YOLO

# 🧠 3. Güçlü model tanımla (yolov8s = small, balanced, good accuracy)
model = YOLO("yolov8s.yaml")  # Alternatif: "yolov8m.yaml" daha büyük, "yolov8n.yaml" daha hafif

# 🎯 4. Eğitim başlat
model.train(
    data="/content/yemek.v1i.yolov8/data.yaml",  # Roboflow'dan gelen dataset
    epochs=50,                 # Yeterince uzun bir eğitim
    imgsz=640,                  # Standart giriş boyutu
    batch=16,                   # Uygun batch size (donanımına göre değişebilir)
    name="yemek_model_gelistirilmis",  # Çıktı klasörü adı
    optimizer="Adam",           # Daha hızlı ve kararlı öğrenme
    lr0=0.001,                  # Başlangıç öğrenme oranı
    weight_decay=0.0005,        # Overfitting'i önler
    augment=True,               # Veri artırma açık (çevirme, yakınlaştırma vs.)
    verbose=True                # Eğitim sırasında detaylı çıktı ver
)
```
