```python
import os
import face_recognition

garson_klasoru = "garson_fotograflari"
garson_encodingleri = []
garson_isimleri = []

# Garson yüzlerini oku
for dosya in os.listdir(garson_klasoru):
    if dosya.endswith(".jpg") or dosya.endswith(".png"):
        yol = os.path.join(garson_klasoru, dosya)
        img = face_recognition.load_image_file(yol)
        enc = face_recognition.face_encodings(img)
        if enc:
            garson_encodingleri.append(enc[0])
            garson_isimleri.append(os.path.splitext(dosya)[0])

# Test fotoğrafı tanı
def tanimla_garson(goruntu_yolu):
    img = face_recognition.load_image_file(goruntu_yolu)
    enc = face_recognition.face_encodings(img)
    if not enc:
        return "Yüz bulunamadı"
    mesafe = face_recognition.face_distance(garson_encodingleri, enc[0])
    en_yakin = mesafe.argmin()
    return garson_isimleri[en_yakin] if mesafe[en_yakin] < 0.6 else "Eşleşme yok"

# Örnek kullanım
print(tanimla_garson("test.png"))
```

```python
import os
import face_recognition

GARSON_FOLDER = "garson_fotograflari"  # notebook ile aynı dizinde

known_encodings = []
known_names = []

for fn in os.listdir(GARSON_FOLDER):
    if fn.lower().endswith(('.png', '.jpg', '.jpeg')):
        path = os.path.join(GARSON_FOLDER, fn)
        img = face_recognition.load_image_file(path)
        encs = face_recognition.face_encodings(img)
        if encs:
            known_encodings.append(encs[0])
            known_names.append(os.path.splitext(fn)[0])
            print(f"✅ {fn} eklendi.")
        else:
            print(f"⚠️ Yüz bulunamadı: {fn}")
```

```python
%pip install numpy

import numpy as np
import face_recognition
import cv2

def recognize_faces_in_frame(frame_bgr, tolerance=0.5):
    """
    frame_bgr: OpenCV'den gelen BGR frame
    tolerance: yüz benzerliği eşiği (0.4 daha sıkı, 0.6 daha gevşek)
    """
    rgb = cv2.cvtColor(frame_bgr, cv2.COLOR_BGR2RGB)
    locs = face_recognition.face_locations(rgb)
    encs = face_recognition.face_encodings(rgb, locs)

    detections = []
    for (top, right, bottom, left), enc in zip(locs, encs):
        dists = face_recognition.face_distance(known_encodings, enc)
        if len(dists) == 0:
            name = "unknown"
        else:
            i = np.argmin(dists)
            name = known_names[i] if dists[i] < tolerance else "unknown"
        detections.append((name, (left, top, right, bottom)))
    return detections
```

```python
def crop_roi(frame_bgr, roi):
    h, w = frame_bgr.shape[:2]
    x1 = int(roi[0] * w)
    y1 = int(roi[1] * h)
    x2 = int(roi[2] * w)
    y2 = int(roi[3] * h)
    return frame_bgr[y1:y2, x1:x2], (x1, y1)  # crop ve orijine offset
```

```python
import pandas as pd
import time
from datetime import datetime

loglar = []

def kayit_ekle(garson_adi, zaman_str):
    loglar.append({
        "Zaman": zaman_str,
        "Garson": garson_adi
    })

def process_video_for_garson(video_path, roi=(0,0,1,1), tolerance=0.5, min_frames_confirm=3, show=True, save_path=None):
    cap = cv2.VideoCapture(video_path)
    if not cap.isOpened():
        print("Video açılamadı:", video_path)
        return

    frame_idx = 0
    seen_counts = {}
    confirmed_garsons = []

    out_writer = None
    if save_path:
        fourcc = cv2.VideoWriter_fourcc(*'mp4v')
        fps = cap.get(cv2.CAP_PROP_FPS) or 25
        frame_interval = int(fps * frame_skip_seconds)  # eklenecek

        w = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
        h = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
        out_writer = cv2.VideoWriter(save_path, fourcc, fps, (w, h))

    while True:
        ret, frame = cap.read()
        if not ret:
            break

        frame_idx += 1
        roi_frame, offset = crop_roi(frame, roi)
        dets = recognize_faces_in_frame(roi_frame, tolerance=tolerance)

        for name, (l, t, r, b) in dets:
            l += offset[0]; r += offset[0]
            t += offset[1]; b += offset[1]
            color = (0,255,0) if name != "unknown" else (0,0,255)
            cv2.rectangle(frame, (l,t), (r,b), color, 2)
            cv2.putText(frame, name, (l, t-10), cv2.FONT_HERSHEY_SIMPLEX, 0.8, color, 2)

            if name != "unknown":
                seen_counts[name] = seen_counts.get(name, 0) + 1
                if seen_counts[name] == min_frames_confirm:
                    ts = time.time()
                    ts_str = time.strftime('%Y-%m-%d %H:%M:%S', time.localtime(ts))
                    confirmed_garsons.append((ts_str, name))
                    kayit_ekle(name, ts_str)
                    print(f"[{ts_str}] Garson görüldü: {name}")

        if show:
            cv2.imshow("garson video", frame)
            if cv2.waitKey(1) & 0xFF == ord('q'):
                break

        if out_writer:
            out_writer.write(frame)

    cap.release()
    if out_writer:
        out_writer.release()
    cv2.destroyAllWindows()

    # Günlük raporu CSV'ye kaydet
    if loglar:
        df = pd.DataFrame(loglar)
        df.to_csv("gunluk_garson_logu.csv", index=False)
        print("📁 'gunluk_garson_logu.csv' dosyası oluşturuldu.")

    return confirmed_garsons
```

```python
VIDEO = "garson_deneme2.mov"  # aynı dizinde
# koridorun sağ yarısı ROI, gerekirse değiştir
confirmed = process_video_for_garson(VIDEO, roi=(0.5,0,1,1), tolerance=0.5, min_frames_confirm=3, show=True, save_path="garson_out.mp4")
print("Tespitler:", confirmed)
```

```python
import ctypes
ctypes.cdll.LoadLibrary("/opt/homebrew/lib/libzbar.dylib")

import os
os.environ["DYLD_LIBRARY_PATH"] = "/opt/homebrew/lib"

import cv2
import time
import csv
from datetime import datetime
from ultralytics import YOLO
import face_recognition
import numpy as np
import pyzbar.pyzbar as pyzbar
from matplotlib import pyplot as plt
from IPython.display import clear_output

# 📦 Eğitimli modeli yükle
yolo_model = YOLO("food_yolo_best.pt")
garson_faces = {}

def load_garson_faces(klasor="garson_fotograflari"):
    if not os.path.exists(klasor):
        print(f"❌ Klasör bulunamadı: {klasor}")
        return
    for dosya in os.listdir(klasor):
        if dosya.endswith((".jpg", ".png")):
            path = os.path.join(klasor, dosya)
            img = face_recognition.load_image_file(path)
            encs = face_recognition.face_encodings(img)
            if encs:
                garson_faces[os.path.splitext(dosya)[0]] = encs[0]
                print(f"👤 Garson yüklendi: {dosya}")
            else:
                print(f"⚠️ Yüz bulunamadı: {dosya}")
    print(f"📦 Toplam yüklenen garson sayısı: {len(garson_faces)}")

def recognize_garson(frame, tolerance=0.5):
    if frame is None or not isinstance(frame, np.ndarray) or frame.size == 0:
        return "unknown", None

    try:
        rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        if rgb is None or rgb.size == 0:
            return "unknown", None

        faces = face_recognition.face_locations(rgb)
        encodings = face_recognition.face_encodings(rgb, faces)

        for (top, right, bottom, left), enc in zip(faces, encodings):
            for name, ref_enc in garson_faces.items():
                match = face_recognition.compare_faces([ref_enc], enc, tolerance)[0]
                if match:
                    return name, (left, top, right, bottom)

        return "unknown", None

    except Exception as e:
        print(f"⚠️ Garson tanıma hatası: {e}")
        return "unknown", None


def detect_qr(frame):
    qr_detector = cv2.QRCodeDetector()
    data, bbox, _ = qr_detector.detectAndDecode(frame)
    if bbox is not None and data:
        return data  # QR kod içeriği (örneğin masa numarası)
    return None

def detect_foods(frame):
    if frame is None or not isinstance(frame, np.ndarray) or frame.size == 0:
        return []

    results = yolo_model(frame, conf=0.4)[0]
    print("🍽️ YOLO kutuları:", results.boxes.data.tolist())  # 🔧 Doğru yere koyduk
    print("🍽️ YOLO tespit sayısı:", len(results.boxes))

    items = []
    for r in results.boxes.data.tolist():
        x1, y1, x2, y2, conf, cls = r
        label = yolo_model.names[int(cls)]
        items.append(label)
    return items





def process_video(video_path, save_path=None, show=True, frame_skip_seconds=0):
    cap = cv2.VideoCapture(video_path)
    if not cap.isOpened():
        print("❌ Video açılamadı!")
        return

    load_garson_faces()
    loglar = []

    fps = cap.get(cv2.CAP_PROP_FPS) or 25
    frame_interval = int(fps * frame_skip_seconds)

    # İlk kareyi alıp boyut öğren
    temp_cap = cv2.VideoCapture(video_path)
    ret, frame = temp_cap.read()
    temp_cap.release()

    if not ret:
        print("❌ İlk kare alınamadı, video bozuk olabilir.")
        return

    frame_height, frame_width = frame.shape[:2]
    fourcc = cv2.VideoWriter_fourcc(*"mp4v")
    out_writer = cv2.VideoWriter(save_path, fourcc, fps, (frame_width, frame_height))

    os.makedirs("yuz_tespit_edilemedi", exist_ok=True)

    frame_id = 0
    aktif_masa = None
    aktif_garson = None

    while True:
        ret, frame = cap.read()
        if not ret or frame is None or not isinstance(frame, np.ndarray) or frame.size == 0:
            break
        frame_id += 1

        # HER KAREYİ İŞLE (frame_interval varsa, buraya ekleyebilirsin)
        # if frame_id % frame_interval != 0:
        #     continue

        frame_copy = frame.copy()
        if frame_copy is None or not isinstance(frame_copy, np.ndarray) or frame_copy.size == 0:
            continue

        # MASA TESPİT
        masa = detect_qr(frame)
        if masa:
            aktif_masa = masa
        else:
            masa = aktif_masa if aktif_masa else "Bilinmiyor"

        # GARSON TESPİT
        garson, bbox = recognize_garson(frame)
        if garson != "unknown":
            aktif_garson = garson
        else:
            garson = aktif_garson if aktif_garson else "unknown"

        # YÜZ KUTUSU
        if bbox:
            l, t, r, b = bbox
            cv2.rectangle(frame, (l, t), (r, b), (0, 255, 0), 2)
            cv2.putText(frame, garson, (l, t - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0, 255, 0), 2)
        else:
            if garson == "unknown":
                cv2.imwrite(f"yuz_tespit_edilemedi/frame_{frame_id:04}.jpg", frame)

        # 🍽️ YEMEK TESPİTİ + KUTU
        try:
            yemekler = detect_foods(frame_copy)
            fiyatlar = {
                "dessert": 300,
                "salad": 200,
                "grilled": 450,
                "soup": 150,
                "stews": 250
            }
            hesap = sum(fiyatlar.get(yemek, 0) for yemek in yemekler)

            results = yolo_model(frame_copy, conf=0.4)[0]
            for r in results.boxes.data.tolist():
                x1, y1, x2, y2, conf, cls = r
                label = yolo_model.names[int(cls)]
                cv2.rectangle(frame, (int(x1), int(y1)), (int(x2), int(y2)), (0, 0, 255), 2)
                cv2.putText(frame, label, (int(x1), int(y1) - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 0, 255), 2)
        except Exception as e:
            print(f"⚠️ YOLO tespiti hatası (frame {frame_id}): {e}")
            yemekler = []
            hesap = 0

        # METİN OVERLAY
        cv2.putText(frame, f"Masa: {masa}", (20, 40), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0, 255, 255), 2)
        cv2.putText(frame, f"Garson: {garson}", (20, 80), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (255, 255, 255), 2)
        cv2.putText(frame, f"Yemekler: {', '.join(yemekler)}", (20, 120), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 0, 255), 2)
        cv2.putText(frame, f"Hesap: ₺{hesap}", (20, 160), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0, 0, 255), 2)
        cv2.putText(frame, f"Kare: {frame_id}", (20, 200), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (200, 100, 255), 2)

        # LOG KAYDI
        now = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
        loglar.append({
            "Zaman": now,
            "Garson": garson,
            "Masa": masa,
            "Yemekler": yemekler
        })

        if show:
            clear_output(wait=True)
            plt.imshow(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))
            plt.title(f"Frame: {frame_id}")
            plt.axis("off")
            plt.show()
            time.sleep(0.05)

        out_writer.write(frame)

    cap.release()
    out_writer.release()

    with open("loglar.csv", "w", newline="", encoding="utf-8") as f:
        writer = csv.DictWriter(f, fieldnames=["Zaman", "Garson", "Masa", "Yemekler"])
        writer.writeheader()
        for row in loglar:
            row["Yemekler"] = ", ".join(row["Yemekler"])
            writer.writerow(row)

    print("✅ İşlem tamamlandı. Log kaydedildi → loglar.csv")
    # 🔽 Çalıştır
process_video("deneme.mov", save_path="cikti_video.mp4", show=False, frame_skip_seconds=2)

```
